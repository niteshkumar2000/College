### Question - 2

Histogram of association between shelf position and sugar content of the variables from the shelf

```python
# data['shelf'].unique()
top_shelf = data[(data['shelf'] == 'Top')]
middle_shelf = data[(data['shelf'] == 'Middle')]
bottom_shelf = data[(data['shelf'] == 'Bottom')]

fig, axs = plt.subplots(1, 3)

axs[0].hist(top_shelf['sugars'])
axs[0].set_title('Top Shelf Sugar')

axs[1].hist(middle_shelf['sugars'])
axs[1].set_title('Middle Shelf Sugar')

axs[2].hist(bottom_shelf['sugars'])
axs[2].set_title('Bottom Shelf Sugar')

top_shelf['sugars'].mean(), middle_shelf['sugars'].mean(), bottom_shelf['sugars'].mean() 
```

### Question - 3

From the histogram, 
- All the three shelves has sugar content range from 0-14.
- It looks like middle shelf has most sugar content (avg of 9.6) followed by top (avg of 6.5) and bottom (avg of 5.1). 

Hence the order is
- middle
- top
- bottom

### Question - 5

Scatter plot between calories and carbohydrates

```python
plt.scatter(data['carbo'], data['calories'])
plt.title('Scatter Plot - Calories vs Carbohydrates')
plt.ylabel('Calories')
plt.xlabel('Carbohydrates')
data['carbo'].corr(data['calories'])
```

From the plot, we can see that the value of calories increases as the carbohydrate values increases so we can say there exist a positive correlation between them but the relationship is weak because the correlation value is very close to zero (i.e. 0.25 in our case). <br>
Hence, we can conclude that these two variables has positive correlation with weak relationship.
